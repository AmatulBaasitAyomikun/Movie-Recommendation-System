```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_data = pd.read_csv('/content/movies.csv')
movie_data.head()
```

```python
# Getting the number of rows and columns

movie_data.shape
```

```python
# Feature Selection

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Replacing null values with null string

# Filter selected_features to include only columns present in movie_data
actual_features_to_process = [feature for feature in selected_features if feature in movie_data.columns]

for feature in actual_features_to_process:
  movie_data[feature] = movie_data[feature].fillna(' ')
```

```python
# combining the selected features
# Using actual_features_to_process which only contains features present in the DataFrame
if actual_features_to_process:
    # Initialize combined_features with the first existing feature
    combined_features = movie_data[actual_features_to_process[0]].astype(str)
    # Concatenate remaining existing features with a space
    for i in range(1, len(actual_features_to_process)):
        combined_features += ' ' + movie_data[actual_features_to_process[i]].astype(str)
else:
    # If no features are available in the DataFrame, create an empty Series
    combined_features = pd.Series([''] * len(movie_data), index=movie_data.index)

print(combined_features)
```

```python
# converting the text to feature vector

vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

Cosine Similarity

```python
# getting the similarity score

similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie_name as input from users

movie_name = input('Enter your favorite movie name: ')
```

```python
# creating a list of movie names from the movie data

all_titles = movie_data['title'].tolist()
print(all_titles)
```

```python
# getting the close match for the user's favorite movie

get_close_match = difflib.get_close_matches(movie_name, all_titles)
print(get_close_match)
```

```python
close_match = get_close_match[0]
print(close_match)
```

```python
# getting the index of the movie name from the dataset

index_of_movie_name = movie_data[movie_data.title == close_match].index.values[0]
print(index_of_movie_name)
```

```python
# getting a list of similar movies based on the similarity score

similarity_score = list(enumerate(similarity[index_of_movie_name]))
print(similarity_score)
```

```python
# sorting the movies based on thier similarity scores

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print similar movies based on index

print('Suggested movies: \n')

i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index == index]['title'].values[0]
  if (i < 31):
    print(i, '.', title_from_index)
    i += 1
```

Movie Recommendation System

```python
movie_name = input('Enter your favorite movie name: ')

all_titles = movie_data['title'].tolist()

get_close_match = difflib.get_close_matches(movie_name, all_titles)

close_match = get_close_match[0]

index_of_movie = movie_data[movie_data.title == close_match].index.values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Suggested movies: \n')

i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index == index]['title'].values[0]
  if (i < 31):
    print(i, '.', title_from_index)
    i += 1
```
